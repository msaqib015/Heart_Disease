```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set(style="whitegrid")
```

```python
df = pd.read_csv('heart.csv')
df
```

```python
print("Dataset shape:", df.shape)
df.head()
```

```python
# Age Distribution by Heart Disease
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='age', hue='target', kde=True, palette='Set1', bins=30)
plt.title('Age Distribution by Heart Disease Presence')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend(title='Heart Disease', labels=['No', 'Yes'])
plt.show()
```

```python
# Sex vs Heart Disease
plt.figure(figsize=(6, 5))
sns.countplot(data=df, x='sex', hue='target', palette='Set2')
plt.title('Sex vs Heart Disease')
plt.xticks([0, 1], ['Female', 'Male'])
plt.xlabel('Sex')
plt.ylabel('Count')
plt.legend(title='Heart Disease', labels=['No', 'Yes'])
plt.show()
```

```python
# Chest Pain Type vs Heart Disease
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='cp', hue='target', palette='Set3')
plt.title('Chest Pain Type vs Heart Disease')
plt.xlabel('Chest Pain Type')
plt.ylabel('Count')
plt.legend(title='Heart Disease', labels=['No', 'Yes'])
plt.show()
```

```python
# Cholesterol Boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='target', y='chol', palette='coolwarm')
plt.title('Cholesterol by Heart Disease')
plt.xlabel('Heart Disease')
plt.ylabel('Cholesterol (mg/dl)')
plt.show()
```

```python
# Resting Blood Pressure Boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='target', y='trestbps', palette='coolwarm')
plt.title('Resting Blood Pressure by Heart Disease')
plt.xlabel('Heart Disease')
plt.ylabel('Resting BP (mm Hg)')
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(12, 10))
corr = df.corr()
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
